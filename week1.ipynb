### Collaborative Coding with Git: Week1

> * Installing Git [link]()

#### Cloning a repository:

> * In this step you will learn how to use Git to make a copy of a codebase using the “clone” command. 

> * Use the cd command to navigate to the directory where you would like the repository to be cloned to. Now you can issue the clone command. The only information git requires is the URI of the repository to be cloned.

**Command:**
$ git clone [link](https://gitlab.com/collaborative-coding-with-git/meetux-simple-example.git)

#### Exploring a cloned repository with the git graph in the terminal: 
> * In the previous step you cloned a remote repository. This means that you now have a local repository which you can use to explore the codebase.
> *  You can use the terminal to display your local repository’s graph by navigating to the project directory and issuing the following command.

**Command:** $ git log --all --oneline --graph

Git vertical(bottom-to-top) graph view: 

> * Fig1: Looking at the graph, you can see that there are three branches in total [add-waiting-room], [add-clothes-shop] and [master] (the other two branches branch-off from master). Our current position on the graph is denoted by the HEAD. Currently, the HEAD is pointing to the most recent commit in [master]. In the terminal, this is depicted by the word HEAD and an arrow pointing to the current commit.

Git horizontal graphs view: 

#### You can switch to a different branch using the switch command.
**Command:** $ git switch add-clothes-shop
> * Using the switch command, we can easily switch back and forth between the branches at will.

**Command:** $ git log --all --oneline --graph
> * If you were to check the graph again, now, you would see that the HEAD has moved on the graph, indicating our position has changed.
